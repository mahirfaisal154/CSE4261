```python
from ultralytics import YOLO
import torch 
import cv2
import os
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
device="cuda"
print(f"{device}")
model=YOLO(r"F:\YOLO\best_widerface_yolov8.pt")
def procerss_video(video_path,output_path):
    cap=cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return 
    #get video properties 
    width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps=int(cap.get(cv2.CAP_PROP_FPS))
    total_frames=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    #define the codec amd create  VideoWriter object
    codecSelection=cv2.VideoWriter_fourcc(*"mp4v")
    out=cv2.VideoWriter(output_path,codecSelection,fps,(width,height))
    if not out.isOpened():
        print("VideoWriter failed to open.")
        return
    print(f"processing video")
    frame_count=0
    while cap.isOpened():
        print(f"{width},{height},{fps}")
        read,frame=cap.read()
        if not read:
            break
        frame_rgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        
        #RUN_INFERENCE
        results=model.predict(source=frame_rgb,conf=.1,iou=.7)
        annoted_frame=results[0].plot()
        annoted_frame=cv2.cvtColor(annoted_frame,cv2.COLOR_RGB2BGR)
        out.write(annoted_frame)
        frame_count+=1
        print(f"processed_frame{frame_count}/{total_frames}")
    cap.release()
    out.release()
    cv2.destroyAllWindows()
    print(f"output saved to path{output_path}")



        
        
    
    


    
```

```python
video_path=r"F:\YOLO\face_video.mp4"
output_path=r"F:\YOLO\face_outpput_finetuned_colab.mp4"
procerss_video(video_path,output_path)
```
